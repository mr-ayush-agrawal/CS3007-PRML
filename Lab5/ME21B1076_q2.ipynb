```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
query_img=cv2.imread('queryimage.png')
ref1_img=cv2.imread('refimage1.png')
ref2_img=cv2.imread('refimage2.png')
```

```python
plt.figure(figsize=(20, 5))
plt.subplot(1, 3, 1)
plt.hist(query_img.ravel(), 256, [0, 256])
plt.title("Query image Histogram")
plt.subplot(1, 3, 2)
plt.hist(ref1_img.ravel(), 256, [0, 256])
plt.title("Refernce 1 Image Histogram")
plt.subplot(1, 3, 3)
plt.hist(ref2_img.ravel(),256,[0,256])
plt.title("Reference 2 Image Histogram")
plt.show()

```

```python
hist_query=cv2.calcHist([query_img],[0,1,2],None,[8,8,8],[0,256,0,256,0,256])
hist_query_norm=cv2.normalize(hist_query,hist_query).flatten()
hist_ref1=cv2.calcHist([ref1_img],[0,1,2],None,[8,8,8],[0,256,0,256,0,256])
hist_ref1_norm=cv2.normalize(hist_ref1,hist_ref1).flatten()
hist_ref2=cv2.calcHist([ref2_img],[0,1,2],None,[8,8,8],[0,256,0,256,0,256])
hist_ref2_norm=cv2.normalize(hist_ref2,hist_ref2).flatten()
dist_query_ref1=cv2.compareHist(hist_query_norm,hist_ref1_norm,cv2.HISTCMP_BHATTACHARYYA)
dist_query_ref2=cv2.compareHist(hist_query_norm,hist_ref2_norm,cv2.HISTCMP_BHATTACHARYYA)
print(dist_query_ref1)
print(dist_query_ref2)
```
